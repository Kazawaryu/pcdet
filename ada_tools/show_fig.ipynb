```python
import seaborn as sns
import os
source_path = './iter_dist/'

A_list = ['A_S2(A_S2)', 'B_S2(A_S2)', 'C_S2(A_S2)', 'D_S2(A_S2)']
B_list = ['A_S2(B_S2)', 'B_S2(B_S2)', 'C_S2(B_S2)', 'D_S2(B_S2)']
C_list = ['A_S2(C_S2)', 'B_S2(C_S2)', 'C_S2(C_S2)', 'D_S2(C_S2)']
D_list = ['A_S2(D_S2)', 'B_S2(D_S2)', 'C_S2(D_S2)', 'D_S2(D_S2)']

a_idx, b_idx, c_idx, d_idx = [[],[],[],[]], [[],[],[],[]], [[],[],[],[]], [[],[],[],[]]
a_num, b_num, c_num, d_num = [[],[],[],[]], [[],[],[],[]], [[],[],[],[]], [[],[],[],[]]

for i in range(len(A_list)):
    path = source_path + A_list[i] + '.txt'
    with open(path, 'r') as f:
        for line in f:
            a_idx[i].append(int(line.split()[0]))
            a_num[i].append(float(line.split()[1]))

for i in range(len(B_list)):
    path = source_path + B_list[i] + '.txt'
    with open(path, 'r') as f:
        for line in f:
            b_idx[i].append(int(line.split()[0]))
            b_num[i].append(float(line.split()[1]))

for i in range(len(C_list)):
    path = source_path + C_list[i] + '.txt'
    with open(path, 'r') as f:
        for line in f:
            c_idx[i].append(int(line.split()[0]))
            c_num[i].append(float(line.split()[1]))

for i in range(len(D_list)):
    path = source_path + D_list[i] + '.txt'
    with open(path, 'r') as f:
        for line in f:
            d_idx[i].append(int(line.split()[0]))
            d_num[i].append(float(line.split()[1]))


res = []
for i in range(len(A_list)):
    res.append(sum(a_num[i])/len(a_num[i]))
    print(A_list[i], res[i])
print(sum(res)/len(res))
print("="*40)
res = []
for i in range(len(B_list)):
    res.append(sum(b_num[i])/len(b_num[i]))
    print(B_list[i], res[i])
print(sum(res)/len(res))
print("="*40)
res = []
for i in range(len(C_list)):
    res.append(sum(c_num[i])/len(c_num[i]))
    print(C_list[i], res[i])
print(sum(res)/len(res))
print("="*40)
res = []
for i in range(len(D_list)):
    res.append(sum(d_num[i])/len(d_num[i]))
    print(D_list[i], res[i])
print(sum(res)/len(res))
print("="*40)
# sns.distplot(a_num, kde=True, rug=False)
# sns.distplot(b_num, kde=False, rug=True)
# sns.distplot(c_num, kde=False, rug=True)
# sns.distplot(d_num, kde=False, rug=True)
```
