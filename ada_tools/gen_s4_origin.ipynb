```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import random

main_dataset_path = '/home/jiazx_ug/dataset/S2'
sub_dirs = ['A','B','C','D']

new_dataset_save_path = '/home/jiazx_ug/dataset/S4/O/'

calib_save_path = new_dataset_save_path+'/training/calib'
image_2_save_path = new_dataset_save_path+'/training/image_2'
label_2_save_path = new_dataset_save_path+'/training/label_2'
velodyne_save_path = new_dataset_save_path+'/training/velodyne'

calib_test_path = new_dataset_save_path+'/testing/calib'
image_2_test_path = new_dataset_save_path+'/testing/image_2'
label_2_test_path = new_dataset_save_path+'/testing/label_2'
velodyne_test_path = new_dataset_save_path+'/testing/velodyne'

imageset_save_path = new_dataset_save_path+'/ImageSets'

if not os.path.exists(calib_save_path):
    os.makedirs(calib_save_path) 
if not os.path.exists(image_2_save_path):
    os.makedirs(image_2_save_path)
if not os.path.exists(label_2_save_path):
    os.makedirs(label_2_save_path)
if not os.path.exists(velodyne_save_path):
    os.makedirs(velodyne_save_path)
if not os.path.exists(calib_test_path):
    os.makedirs(calib_test_path)
if not os.path.exists(image_2_test_path):
    os.makedirs(image_2_test_path)
if not os.path.exists(label_2_test_path):
    os.makedirs(label_2_test_path)
if not os.path.exists(velodyne_test_path):
    os.makedirs(velodyne_test_path)
if not os.path.exists(imageset_save_path):
    os.makedirs(imageset_save_path)
```

```python

data_idxs=[[],[],[],[]]

for i in range(len(sub_dirs)):
    sub_dir = sub_dirs[i]
    imageset_path = main_dataset_path+'/'+sub_dir+'/ImageSets/train.txt'
    with open(imageset_path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            data_idxs[i].append(line.strip())
```

```python
for i in range(len(data_idxs)):
    if i == 0:
        ture_main_set_name = 'A'
    elif i == 1:
        ture_main_set_name = 'B'    
    elif i == 2:
        ture_main_set_name = 'C'
    else:
        ture_main_set_name = 'D'

    for j in range(len(data_idxs[i])):
        data_idx = data_idxs[i][j]
        calib_path = main_dataset_path+'/'+ture_main_set_name+'/training/calib/'+data_idx+'.txt'
        image_2_path = main_dataset_path+'/'+ture_main_set_name+'/training/image_2/'+data_idx+'.png'
        label_2_path = main_dataset_path+'/'+ture_main_set_name+'/training/label_2/'+data_idx+'.txt'
        velodyne_path = main_dataset_path+'/'+ture_main_set_name+'/training/velodyne/'+data_idx+'.bin'

        calib_save_path = new_dataset_save_path+'/training/calib/'+data_idx+'.txt'
        image_2_save_path = new_dataset_save_path+'/training/image_2/'+data_idx+'.png'
        label_2_save_path = new_dataset_save_path+'/training/label_2/'+data_idx+'.txt'
        velodyne_save_path = new_dataset_save_path+'/training/velodyne/'+data_idx+'.bin'

        os.system('cp '+calib_path+' '+calib_save_path)
        os.system('cp '+image_2_path+' '+image_2_save_path)
        os.system('cp '+label_2_path+' '+label_2_save_path)
        os.system('cp '+velodyne_path+' '+velodyne_save_path)

        print('copy', data_idx, 'done')


```

```python
# combine data_idxs into one list
str_idx_list = data_idxs[0] + data_idxs[1] + data_idxs[2] + data_idxs[3]
str_idx_list = list(set(str_idx_list))


val_idx_list = np.random.choice(str_idx_list, int(len(str_idx_list)*0.1), replace=False)
train_idx_list = list(set(str_idx_list) - set(val_idx_list))

val_idx_list = [elem for elem in val_idx_list]
train_idx_list = [elem for elem in train_idx_list]

val_idx_list = list(set(val_idx_list))
train_idx_list = list(set(train_idx_list))

imagesets_train_path = new_dataset_save_path + '/ImageSets/train.txt'
imagesets_val_path = new_dataset_save_path + '/ImageSets/val.txt'
imagesets_test_path = new_dataset_save_path + '/ImageSets/test.txt'


with open(imagesets_train_path, 'w') as f:
    for idx in train_idx_list:
        f.write(idx+'\n')
with open(imagesets_val_path, 'w') as f:
    for idx in val_idx_list:
        f.write(idx+'\n')
with open(imagesets_test_path, 'w') as f:
    pass
```
