```python
import os
import pandas as pd
import numpy as np

old_path = 'pointpillar_copy2_20_0.7_hard_result.txt'
new_path = 'pointpillar_copy2_30_0.7_hard_result.txt'

old = []
new = []

with open(old_path, 'r') as f:
    for line in f:
        idx, rap = line.strip().split()
        old.append((idx, rap))

with open(new_path, 'r') as f:
    for line in f:
        idx, rap = line.strip().split()
        new.append((idx, rap))

# sort by idx
old = sorted(old, key=lambda x: str(x[0]))
new = sorted(new, key=lambda x: str(x[0]))
```

```python
# use new to minus old value
diff = []
for i in range(len(old)):
    idx = old[i][0]
    old_rap = float(old[i][1])
    new_rap = float(new[i][1])
    
    diff.append((idx, (new_rap - old_rap)))

# sort by diff
diff = sorted(diff, key=lambda x: x[1], reverse=True)
diff
```

```python

# check the len of diff which value < 0
count = 0
idx_select = []

# drop top 1000
# for i in range(1000, len(diff)):
#     idx_select.append(diff[i][0])


for i in range(len(diff)):
    if diff[i][1] < 0 or (diff[i][1] == 0 and float(old[i][1]) > 0.5):
        count += 1
    else:
        idx_select.append(diff[i][0])
print(len(idx_select))
save_path = './../idxs_s4/train.txt'

with open(save_path, 'w') as f:
    for idx in idx_select:
        f.write(idx + '\n')

save_path = './../idxs_s4/val.txt'

with open(save_path, 'w') as f:
    for idx in idx_select:
        f.write(idx + '\n')
```
